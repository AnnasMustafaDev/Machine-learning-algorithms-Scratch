```python
import numpy as np
```

```python
x=np.array([94,88,75,71,75,68,73,57,59,65,67,60,60,45,61])
y=np.array([80,60,56,47,43,40,67,16,21,54,40,39,38,8,34])
```

```python
print(y.size)
x.shape
```

```python
print('x mean : ',x.mean())
a=x.mean()                       # x` 
print('y mean : ',y.mean())
b=y.mean()                       # y`
```

```python
q1=x-a #x-x`
q2=y-b #y-y`
q=q1*q2
```

```python
cov1=np.sum(q) 
cov1
cov=cov1/(x.size-1) 
cov   #covariance  Σ(x-x`)(y-y`)/n
```

```python
q1
w1=q1**2  #(x-x`)^2
w=np.sum(w1)
w
var= w/(x.size-1)
var #variance  Σ(x-x`)^2/n
```

```python
# y = b0 + (b1)(x)
# y , dependant variable
# x , independant variable
# b0 , intercept
# b1 , slope

b1=cov/var
print('Slope ~ Coefficeint :',b1)
```

```python
# b0 = y`- b1 x`
print(b)
print(a)
print(b1)
b0=b-(b1*a)
```

```python
b0
```

```python
ye=np.array([])
for i in range(15):
    print('predicted y for x',i+1,': ',b0+(b1*x[i]))
    element = b0+(b1*x[i])
    ye = np.append(ye, element)
```

```python
#ye= yPredicted
ye.size
ye
```

### Coefficient of Corelation

```python
# r = cov/ std(x)std(y)
# std(x)=  sqrt  (Σ(x-x`)/n)
```

### Coefficient of determination

```python
# Coefficient of Determination
#R^2 = Σ(ye-y`)^2/ Σ(y-y`)^2
```

```python
# root mean square error RMSE= sqrt (Σ(ye-y)/(x.size-1))
rm=(ye-y)**2
```

```python
rmse=np.sum(rm)/15
print('Mean sq Error : ',rmse)
print('Root mean sq Error : ',np.sqrt(rmse))
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y,ye)
```
